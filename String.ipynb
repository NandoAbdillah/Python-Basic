# String


```python
# String dapat didefinisikan dengan " " , ' ', """ """, ''' ''' . Tergantung anda
# Indeksing dalam string sama dengan list setiap karakter masuk ke dalam 1 indeks termasuk huruf,karakter,spasi
# Perhatikan contoh dibawah ini

greetings = "Good Morning! Have a nice day"
```

```python
for i in range(0,len(greetings)):
    print(f" indeks ke {i} => {greetings[i]}")
```

```python
#Berdasarkan diatas setiap karakter tersimpan di indeks tertentu termasuk spasi dan karakter khusus
```
