# Fungsi Rekursif
##### Fungsi rekursif adalah sebuah fungsi unik yang membentuk sebuah perulangan dengan pola tertentu
##### Perhatikan contoh fungsi rekursif dibawah ini


```python
# Contoh Faktorial menggunakan fungsi rekursif

def factorial (x) :
    if x==1 :
        return 1;
    else :
        return x* factorial(x-1)
```

```python
factorial (5)
```

```python
# Contoh menambahkan bilangan
# 8 -> 8+7+6+5+4+3+2+1

def adding (x) :
    if x==1 :
        return 1
    else :
        return x + adding(x-1)
```

```python
adding(5)
```

```python
# Contoh bilangan Fibonacci 
# 0 1 1 2 3 5 8 13 dst....

def fibonacci (x) :
    if x == 1:
        return 0
    elif x==2:
        return 1
    else :
        return (fibonacci(x-1) + fibonacci(x-2))

x = int(input("Input no for Fibonacci : "))
list = []
for i in range(1,x+1) :
    list.append(fibonacci(i))

str_ = ""

for i in list :
    str_ += " "+ str(i)
print (str_)

```
