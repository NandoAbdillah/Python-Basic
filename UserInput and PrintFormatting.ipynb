```python
# Skenario
# 1. Andi harus menginputkan umurnya dalam bentuk integer
# 2. Kakak andi lahir pada tahun 1996
# 3. Sekarang tahun 2024

# Output yang diharapkan
# "Sekarang tahun 2024, Andi berusia ... tahun sedangkan kakaknya berusia ... tahun, Kakaknya ... tahun lebih tua dari Andi yang lahir di tahun ..."
```

```python
umurAndi = int(input("Andi:: Masukkan umur kamu :"))
```

```python
umurAndi
```

```python
print("Sekarang tahun 2024, Andi berusia ", umurAndi," tahun sedangkan kakaknya berusia ",2024-1996," tahun, kakaknya ", 2024-1996-umurAndi, "tahun lebih tua dari Andi yang lahir di tahun ",2024-umurAndi) 
```

```python
# Namun ada 2 cara lain agar penulisan string dan variabel lebih mudah/efisien
```

```python
umurKakak = 2024-1996
```

```python
tahunLahirAndi = 2024-umurAndi
```

```python
selisih = umurKakak-umurAndi
```

```python
print("Sekarang tahun 2024, Andi berusia {a} tahun sedangkan kakaknya berusia {b} tahun, kakaknya {c} tahun lebih tua dari Andi yang lahir di tahun {d}".format(a=umurAndi,b=umurKakak,c=selisih,d=tahunLahirAndi))
```
