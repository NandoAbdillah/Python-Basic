# Append
##### Append berfungsi untuk menambahkan 1 nilai ke dalam list 1 nilai tersebut dapat berupa macam-macam tipe data termasuk array
##### PERLU DI INGAT : HANYA MENYISIPKAN 1 NILAI SAJA

```python
nilai_siswa.append(81)
```

```python
nilai_siswa
```

```python
# ketika diappend maka otomatis diposisikan/ditempatkan ke paling belakang list
# Apabila kita memberikan array maka yang ditempatkan juga arraynya bukan angkanya

nilai_siswa.append([87,65,81])
```

```python
nilai_siswa
```

# Extend

##### Extend digunakan untuk menyisipkan lebih dari 1 data , jika yang disisipkan adalah list, maka list tersebut akan dilebur

```python
nilai_siswa.extend([83,97])
```

```python
nilai_siswa
```

```python
nilai_siswa_susulan = [76,83]
```

```python
nilai_siswa.extend(nilai_siswa_susulan)
```

```python
nilai_siswa
```

# Insert
##### Insert digunakan untuk menyisipkan di indeks tertentu

```python
# 2 adalah indeks yang dituju , 88 adalah nilainya, 
nilai_siswa.insert(2,88)
```

```python
nilai_siswa
```

```python
nilai_siswa.insert(-3,74)
```

# POP
##### POP digunakan untuk menghapus 1 nilai array di posisi paling belakang (default)
##### namun kita juga dapat menghapus di indeks yang spesifik

```python
nilai_siswa.pop()
```

```python
nilai_siswa.pop(-6)
```

```python
nilai_siswa.pop(0)
```

```python
nilai_siswa = nilai_siswa[::3]
```

```python
nilai_siswa
```

```python
nilai_siswa.pop(-2)
```

```python
nilai_siswa
```
