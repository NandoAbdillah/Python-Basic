# String Method
##### String Method merupakan fungsi bawaan khusus String yang disediakan oleh python 

# 1. Strip
##### Strip berfungsi untuk menghapus whitespace dari awal hingga akhir dari sebuah string

# Contoh

ex = "      Namaku Nando"

```python
ex.strip()
```

# 2. Lower
##### Lower digunakan untuk mengecilkan/menguppercasekan semua huruf

```python
# Contoh

ex = "IR SOEKARNO adalah presiden RI pertama"
```

```python
ex.lower()
```

# 3. Upper
##### Upper digunakan untuk mengkapitalkan semua huruf

```python
# Contoh 

ex = "perang dunia ke-ii  terjadi diantara tahun 1941-1945"
```

```python
ex.upper()
```

# Replace 
##### Replace digunakan untuk menggantikan kata yang lama ke kata yang baru
##### Syntaknya var.replace(kata yang ingin diubah, kata penggantinya)
##### Ini akan berlaku ke semua kata yang sama

```python
# Contoh

ex = "Ia sangat bahagia ketika melihat kamu bahagia"
```

```python
ex.replace("bahagia","senang")
```

```python
# Ini akan tidak terjadi apa-apa kalau kita tidak menuliskan katanya dengan benar(disini kapital juga berpengaruh)
ex.replace("ia", "dia")
#disini huruf ia dalam bahagia yang malah berubah
```

```python
ex = ex.replace("Ia", "Ayahnya")
ex = ex.replace("kamu","Putra sulungnya")
ex = ex.replace("gembira", "senang")
```

```python
ex
```

# Split 
##### Split digunakan untuk membuat list dari sebuah string dengan syarat-pemisah tertentu
##### syntaks var.split("syarat-pemisah") syarat pemisah bisa berupa spasi atau karakter tertentu

```python
# Contoh

ex = "aku+dan+kamu+ditakdirkan+untuk+bersama"
```

```python
ex.split("+")
```

```python
ex = "     Mereka   bilang  kita kan   berpisah   "
```

```python
ex.split()
```

```python
ex.split(" ")
```
