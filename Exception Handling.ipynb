```python
x = "Hello"

try :
    y = float(x)
    print(y)
except :
    print("Please enter a number")
finally :
    print("Finished....")
```

```python
# Perhatikan kode dibwah ini

try :
    #Mencoba menjalankan kode dibawah ini
    x = input("Masukkan sebuah angka ")
    y = int(x) + 25
    print(y)
except :
    print("Error! Anda harus menginputkan angka")
finally :
    print("Finished")

# Jika di try tidak terjadi error maka bagian except tidak dijalankan dan langsung ke finally
# namun jika di try error maka python menjalankan except dan finally
```

```python
x = 5
y = 8

try :
    print(x+y)
except TypeError :
    print("Both should be numbers")
finally :
    print("Finished...")
```

```python
x = 'Hello'

try :
    print(int(x))
except ValueError :
    print("Enter a number!")
finally :
    print("Finished...")
```

```python
x= 2
y= 0

try :
    print(x/y)
except ZeroDivisionError :
    print("Divided by zero")
finally :
    print("Finished...")
```
