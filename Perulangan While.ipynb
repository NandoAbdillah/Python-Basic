# Perulangan While 
##### While hampir sama dengan for namun ada beberapa yang menjaddi pembedanya

```python
# Contoh penggunaan while

x= 0
y =0

while (x+y < 10) :
    print(x, y)
    x +=1
    y +=2
```

```python
# Faktorial 

angka = int(input("Masukkan angka : "))
fakt =1;
while (angka >= 1) :
    fakt *=angka
    angka-=1
    print(angka)

print(fakt)
```
