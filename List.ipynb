```python
# Membuat List Baru

nama_siswa = ['Nando','Rizky','Zahra','Inul','Khoirul']

# Kita dapat memanggil indeks dari depan dan juga dari belakang
# Contoh dari depan :
# 1. nama_siswa[0] -> Nando
# 2. nama_siswa[1] -> Rizky

#Contoh dari belakang :
# 1. nama_siswa[-1] -> Khoirul
# 2. nama_siswa[-2] -> Inul

```

```python
nama_siswa[-4]
```

```python
nama_siswa[-3]
```

```python
nama_siswa[-2]
```

```python
# Dalam python kita juga dapat meminta panjang list dengan cara len(nama_list)

len(nama_siswa)
```

```python
nama_siswa[len(nama_siswa)-1]
```

```python
# Dari contoh di atas dapat di tarik kesimpulan bahwa sebenarnya memanggil indeks dari belakang menggunakan -bil
# dijabarkan dari panjang list -1
# maka dari itu lebih baik menggunakan list[-bil] jika ingin memanggil list dari belakang
```

# Slice List

```python
# Untuk slice hampir sama dengan slice di js namun ada beberapa perbedannya, seperti
# nama_list[indeks mulai : indeks batas], didalam pemotongan indeks batas tidak boleh dipotong
# jadi apabila list[1:4] maka indeks yang dipotong harus kurang dari indeks 4 yaitu indeks 3
# maka yang dipotong hanya indeks-> 1,2,3
```

```python
nama_siswa
```

```python
nama_siswa[2:4]
```

```python
nama_siswa[1:5]
```

```python
nama_siswa[0:3]
```

```python
# Kita juga bisa hanya mengisi batas awal atau batas akhirnya saja
```

```python
nama_siswa[:2]
```

```python
nama_siswa[3:]
```

```python
# Atau menggunakan indeks dari belakang
```

```python
nama_siswa[1:-1]
```

```python
# Slice 2 Kolon
# nama_list[indeks mulai:indeks batas: banyak langka]
# perhatikan penerapannya dibawah ini

nama_siswa[0::2]
#maka akan melewati 1 indeks di setiap iterasi
```

```python
nama_siswa
```

```python
# Dari Belakang 
#2 langakah
# melewati 1 indeks
nama_siswa[-1::-2]
```

```python
nama_siswa[::-2]
```
