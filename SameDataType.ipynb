# Perbedaan Numpy Array dan List 
##### Numpy array dan list memiliki perbedaan dalam menyimpan anggota arraynya 
##### Jika list , kita boleh mencampurkan semua tipe data dalam satu array tanpa ada konversi paksa , li = [1,True,2.5,"Nama"]
##### di Numpy Array jika tipe data tidak sama maka tipe data akan otomatis dipaksa di konversi ke tipe data
##### jika int,float bertemu string maka akan dipaksa dioonversi ke string
##### penentuan tipe data berdasarkan tipe data mayoritas

### *Numpy Array is Single Data Type*

```python
list = [120,'Nama',12.5,True]
```

```python
import numpy as np

arr = np.array(list)
```

```python
arr # Terkonversi menjadi string
```

```python
arr2 = np.array([1,2,3,4,5,True,False,'String'])
```

```python
arr2
```

```python
arr2 = np.array([1,2,3,4,5,True,False])
# True menjadi 1
# False menjadi 0
```

```python
arr2
```

```python
arr2 = np.array([1,2,3,4,5,True,False,2.5,9.7]) #Terkonversi ke float
```
