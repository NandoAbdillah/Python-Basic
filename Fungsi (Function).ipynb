# Function
#### Function dalam python juga hampir sama dengan bahasa pemrograman yang lainnya
#### Berikut ini contoh penerapan fungsi dalam python

```python
# Contoh

def square (x) :
    area = x**2
    return area
```

```python
square(8)
```

```python
def adding (a,b,c) :
    add = a + b + c
    print("Addition is ->", add)
```

```python
adding(3,5,2)
```

```python
type(adding)
```

```python
type(adding(3,5,2))
```

```python
def adding_multiply (a,b,c) :
    add = a + b + c
    multip = a * b * c
    _add = "Hasil Penjumlahan -> " + str(add)
    _multip = "Hasil Perkalian -> " + str(multip)
    return([_add,_multip])
```

```python
adding_multiply(2,3,4)
```

```python
type(adding_multiply(2,3,4))
```

### Faktorial menggunakan fungsi

```python
def factorial(x):
    hasil = 1
    for i in range(1,x+1):
        hasil *= i

    return hasil
    
```

```python
factorial(5)
```

```python
def sort_array (li) :
    li.sort()
    li.reverse()
    return li
    
```

```python
sort_array([98,76,54,87])
```
