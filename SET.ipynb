# SET
##### SET merupakan suatu tipe data yang mirip mirip dengan list yang membedakannya adalah set menggunakan pengurung {}

```python
# Contoh 

A = {1,2,3,"Sidoarjo",True,178.65}
```

```python
type(A)
```

```python
kota = ["Surabaya","Medan","Jakarta"]
```

```python
# Konversi list ke Set
C = set(kota)
```

```python
C
```

```python
type(kota)
```

```python
# Konfersi Set ke List
D = list(A)
```

```python
D
```

```python
type(D)
```
