# Indexing

```python
import numpy as np

arr = np.array([1,2,3,4,5])
```

```python
len(arr)
```

```python
for i in range (len(arr)):
    print(f" indeks ke {i} -> {arr[i]}")
```

# Negative Indexing

```python
i = -1
while (i != -(len(arr))-1) :
    print(f"Indeks ke {i} -> {arr[i]}")
    i-=1
```

# Slicing 

```python
# Slcing pada array numpy sama dengan list 

arr[:]
```

```python
arr[1:3]
```
