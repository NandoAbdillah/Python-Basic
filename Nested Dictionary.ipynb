# Berikut ini Contoh penggunaan nested dictionary

```python
data = {
    'Nando' : {
        'RPL' : 92,
        'MTK' : 91,
        'ENG' : 90
    },
    'Rayan' : {
        'RPL' : 90,
        'MTK' : 95,
        'ENG' : 89
    },
    'Devano' : {
        'RPL' : 97,
        'MTK' : 88,
        'ENG' : 96
    }
}
```

```python
data
```

```python
# Contoh pemanggilannya
data['Nando']
```

```python
data['Devano']['MTK']
```

```python
# Dengan for

for i in data :
    print(i)
```

```python
for i in data :
    print(i, data[i])
```

```python
for i in data :
    print(f"{i} mendapatkan nilai {data[i]['MTK']} dalam PAS Matematika")
```

```python
# MengEkspor data ke mapel

mapel = [] #Data kosong

for i in data['Nando'] :
    mapel.append(i)
```

```python
mapel
```

```python
# Contoh mrnyaring nilai tertinggi permapel dengan mengekspor nilai ke list

nama = ['Nando','Rayan','Devano']
for i in mapel :
    nilai = [] # List kosong
    for j in data :
        
        nilai.append(data[j][i])
        # print(data[j][i])
    print(f"Nilai {i} tertinggi {max(nilai)} diraih oleh {nama[nilai.index(max(nilai))]} ")
        
```
