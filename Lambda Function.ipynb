# Lambda

```python
# Perhatikan fungsi ini

def square (x) :
    return x*x

# Di atas merupakan fungsi square dengan return
```

```python
square(15)
```

```python
# Itu dapat disederhanakan dengan fungsi lambda yaitu

square_lambda = lambda x: x*x
# variabel_lambda = lambda var : operasi return
```

```python
square_lambda(15)
```

```python
# Beberapa contoh yang lainnya yaitu

def addition (x,y,z) :
    add = x+y+z
    return add
```

```python
addition(5,3,2)
```

```python
# Menggunakan lambda

addition_lambda = lambda x,y,z : x+y+z
```

```python
addition_lambda(5,3,2)
```

```python
# Bagaimana jika ad 2 return , perhatikan contoh fungsi dibawah ini

def greater (x) :
    if x > 15:
        return "x is greater than 15"
    else :
        return "x is lesser than 15"
    
```

```python
greater(10)
```

```python
greater(16)
```

```python
# Menggunakan Lambda untuk kode yang lebih ringkas

greater_lambda = lambda x : "x is greater than 15 " if x>15 else "Lesser"
```

```python
greater_lambda(15)
```

```python
greater_lambda(17)
```

```python
def add(x,y) :
    add = x+y
    if add > 10 :
        return True
    else :
        return False
```

```python
add(5,3)
```

```python
# Lebih ringkas menggunakan lambda

Sumisgreaterthan = lambda x,y : True if x+y>10 else False
```

```python
Sumisgreaterthan(8,9)
```

```python
# Memeriksa kabisat
kabisat = lambda tahun : "Tahun Kabisat" if tahun%4 == 0 else "Bukan Tahun Kabisat"
```

```python
kabisat(1979)
```

```python
kabisat(2024)
```
