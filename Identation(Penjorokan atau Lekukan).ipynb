# Identation 
##### Identation adalah penulisan kode dengan spasi/penjorokan yang dikhususkan untuk beberapa kondisi seperti if,for,while dll
##### hal tersebut dikarenakan syntax python yang tidak menggunakan {} sebagai iddentifikasi statemen
##### Contoh dalam javascript
##### if (logic) {
#####    statement
#####  }
##### Di bawah ini adalah contoh dari penggunaan identation :

```python
#Perhatikan spasi setelah pendeklarasian if
umur = 12
esKrim = 0
if (umur>18):
    #setelah if,for dll harus diberi colo (:)
    esKrim = 1
    
else :
    for i in range(0,5):
        esKrim = i
    esKrim = esKrim+1


print(f"Dapat {esKrim} Es Krim")
    
```
