# Break & Continue 
##### Break & Continue digunakan untuk 
##### Break -> menghentikan looping
##### Continue -> meloncati sebuah loop kemudian loop masih dilanjutkan hingga selesai 


```python
# Contoh penggunaan break 

i = 0

while (i<5):
    if(i==3):
        break
    print(i)
    i+=1
```

```python
# Contoh penggunaan break pada list

list = ['Mangga', 'Melon', 'Apel', 'Kelengkeng']

for i in list :
    if(i=='Melon'):
        list.remove('Melon')
        break
    print(i)
```

```python
# Contoh penggunaan continue pada list
# Menampilkan bilangan prima

for i in range (1,15) :
    if (((i%2==0)|(i%3==0)|(i%5==0)|(i%7==0)) and ((i!=2)&(i!=3)&(i!=5)&(i!=7))) :
        continue
    print(i)
```
