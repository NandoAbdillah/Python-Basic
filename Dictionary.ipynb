# Dictionary
##### Dictionary adalah tipe data yang sama persis dengann array asosiatif di php

```python
# Contoh

siswa_nilai = {'Nando' : 96, 'Andi':88, 'Findi':78}
```

```python
type(siswa_nilai)
```

```python
siswa_nilai['Nando']
```

```python
siswa_nilai['Andi']
```

```python
# Mneampilkan dictionary menggunakan for

for i in siswa_nilai :
    print(i)

# Hanya mengoutputkan key nya saja  , lalu bagimana cara mengoutputkan value nya juga
```

```python
# Cara menampilkan key beserta valuenya 

for i in siswa_nilai :
    print(i,siswa_nilai[i])
```

```python
# Untuk menampilkan valuenya saja

for i in siswa_nilai.values() :
    print(i)
```

```python
#  Cara ke 2 untuk menampilkan key dan value , ini adalah cara yang lebih direkomendasikan

for key,value in siswa_nilai.items() :
    print (key, "->",value)
```

```python
# Value dictionary juga bisa memuat List

nilai = {'Jaka' : [98,86,91], 'Sri':[87,95,85]}
```

```python
nilai['Jaka']
```

```python
# Cara pertama untuk mengoutputkannya
for key in nilai :
    for value in nilai[key]:
        print(key,value)
```

```python
#Cara kedua
for akey,value in nilai.items():
    for li in value :
        print(key, "->", li)
```

# Menambahkan anggota baru

```python
random = {'Nando' : [16,6,2006], '6' : 'Juni', '2006' : str(2024-2006)}
```

```python
# Memasukkan Zovan ke diktionari
random['Zovan'] = [21,6,2007]
```
