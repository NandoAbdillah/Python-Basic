```python
# Perhatikan Penggunaan Map dibawah ini

x = [1,2,3,4,5,6,7,8]

def square (x) :
    return x*x
```

```python
# Saya ingin membuat setiap anggota list x dipangkatkan menggunakan fungsi square
# Alih-alih menggunakan pengulangan untuk setiap elemen, kita lebih baik menggunakan map

result = map(square,x)
#map (fungsi, array)
```

```python
result
```

```python
type(result)
```

```python
# Map sebenarnya berisi kode unik, sehingga kita harus mengcastingnya ke bentuk list
list(result)
```

```python
nama = ("NandO", "AbdI l l ah", "  S a lam", "   bIn ", "N  u r Hu da")
```

```python
def rewrite (txt) :
    txt = txt.replace(" ","")
    txt = txt.lower()
    return txt
```

```python
rewrite(" v s sjf eko")
```

```python
# Menggunakan Map

rewriteNama = map(rewrite,nama)
```

```python
rewriteNama
```

```python
tuple(rewriteNama)
```

```python
# Menjumlahkan 2 list menggunakan Map

def rata_rata (x,y) :
    return (x+y)/2
```

```python
rata_rata(3,2)
```

```python
#Hitunglah rata-rata dari list dibawah ini
#Kolom = rpl, mtk, ing

Andi = [87,77,98]
Gofar = [92, 88, 65]

rata2siswa = map(rata_rata,Andi,Gofar)
```

```python
list(rata2siswa)
```

```python
# Menggunakan Lambda

ganjil_genap = lambda x : "Genap" if x%2 ==0 else "Ganjil"
```

```python
num = [34,33,78,90,42]
hasil = map(ganjil_genap,num)
```
